```python
%%sql
DROP TABLE IF EXISTS tempstudmh;
CREATE TABLE tempstudmh
(
    date_time int
    ,gender varchar(10)
    , age int
    , department varchar (20)
    , study_year int
    , cgpa_score varchar (10)
    , depression varchar (10)
    , anxiety varchar (10)
    , panic_attack varchar (10)
    , specialist_hlp varchar (10)
);

SELECT *
FROM tempstudmh
```

```python
%%sql
INSERT INTO tempstudmh
SELECT date_time, gender, age, department, study_year, cgpa_score, depression, anxiety, panic_attack, specialist_hlp
FROM studmh;

SELECT *
FROM tempstudmh;
```

```python
%%sql
SELECT department, study_year, cgpa_score, count(date_time)
FROM tempstudmh
WHERE depression = 'Y'
AND anxiety = 'Y'
AND panic_attack = 'Y'
GROUP BY department;
```

```python
%%sql
SELECT department, study_year, cgpa_score, count(date_time)
FROM tempstudmh
WHERE (depression = 'Y'
OR anxiety = 'Y'
OR panic_attack = 'Y')
GROUP BY department;
```

Querying anxiety and depression grouped by department shows a trend in students experiencing anxiety and depression with CGPA over 3.00.

```python
%%sql
SELECT *
FROM tempstudmh
WHERE anxiety = 'Y'
AND depression = 'Y'
AND panic_attack = 'Y' 
GROUP BY department
```

Of 101 students, 5 experienced all three concerns depression, anxiety and panic attack, all of which have CGPA of 3.00 > none of which sought specialst help. There is no defining trend regarding department or study year in this sample group.

```python
%%sql
SELECT *
FROM tempstudmh
WHERE specialist_hlp = 'Y';
```

```python
%%sql
SELECT count(*), gender, study_year, cgpa_score
FROM tempstudmh
WHERE anxiety = 'Y'
AND depression = 'Y'
AND panic_attack = 'Y' 
AND specialist_hlp = 'Y'
GROUP BY department
```

Of the 101 students only 1, repied yes to all fields - female, 3rd year student with 3.5-4 CGPA. 

```python
%%sql
SELECT
SUM(CASE WHEN gender = 'M' THEN 1 ELSE 0 END) AS male_count,
  SUM(CASE WHEN gender = 'F' THEN 1 ELSE 0 END) AS female_count
FROM tempstudmh;
```
