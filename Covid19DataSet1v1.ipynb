```python
%%sql
SELECT country, population, MAX(total_confirmed) AS Max_confirmed, total_deaths, CAST((total_confirmed/population)*100 AS float) AS percentpop_confirmed
FROM ctotal
--WHERE country IN ('UK', 'Germany', 'Italy', 'Spain', 'France', 'Greece', 'Sweden', 'Portugal')
GROUP BY country, population
ORDER BY percentpop_confirmed DESC
```

```python
%%sql
SELECT country, population, CAST(total_deaths AS int) AS maxtotal_deaths
FROM ctotal 
--WHERE country IN ('UK', 'Germany', 'Italy', 'Spain', 'France', 'Greece', 'Sweden', 'Portugal')
GROUP BY country, population
ORDER BY maxtotal_deaths DESC
```

```python
%%sql
SELECT continent, MAX(CAST(total_deaths AS int)) AS contt_deaths
FROM ctotal
WHERE continent IS NOT NULL
GROUP BY continent
ORDER BY contt_deaths DESC
```

```python
%%sql
SELECT SUM(max_confirmed) AS sum_confirmed_per_country
FROM (
    SELECT MAX(total_confirmed) AS max_confirmed
    FROM ctotal
    GROUP BY country
) AS max_confirmed_per_country;
```

```python
%%sql
SELECT CAST(SUM(max_confirmed) AS int) AS sum_confirmed_per_country, CAST(SUM(max_deaths) AS int) AS sum_death_per_country, ROUND(CAST((SUM(max_deaths)/SUM(max_confirmed))*100 AS float),1) AS sum_Perct_global_death
FROM (
    SELECT MAX(total_confirmed) AS max_confirmed, MAX(CAST(total_deaths AS float)) AS max_deaths
    FROM ctotal
    GROUP BY country
)
```

Rolling sum for continent

```python
%%sql
SELECT CT.country, CT.continent, CT.population, CD.active_cases, 
SUM(CAST(CD.active_cases AS int)) OVER (PARTITION BY CT.continent ORDER BY CT.continent, CT.country) AS rollingactive_cases
FROM ctotal AS CT
JOIN cdeath AS CD
    ON CT.country = CD.country 
    AND CT.continent = CD.continent
ORDER BY  CT.continent;
```

CTE

```python
%%sql
WITH activecase_continent(country, continent, population, active_cases, rollingactive_cases) AS(
SELECT CT.country, CT.continent, CT.population, CD.active_cases, 
    SUM(CAST(CD.active_cases AS int)) OVER (PARTITION BY CT.continent ORDER BY CT.continent, CT.country) AS rollingactive_cases
FROM ctotal AS CT
JOIN cdeath AS CD
    ON CT.country = CD.country 
    AND CT.continent = CD.continent
ORDER BY  CT.continent)

SELECT *, ROUND(CAST(active_cases AS float)/CAST(population AS float)*100,2) AS percentofrollingpop
FROM activecase_continent
```

TEMP TABLE

```python
%%sql
DROP TABLE IF EXISTS tempactivepop;

CREATE TABLE tempactivepop
(
 country TEXT,
 continent TEXT,
 population REAL,
 active_cases REAL,
 rollingactive_cases REAL
);

INSERT INTO tempactivepop 
SELECT CT.country, CT.continent, CT.population, CD.active_cases, SUM(CAST(CD.active_cases AS int)) OVER (PARTITION BY CT.continent ORDER BY CT.continent, CT.country) AS rollingactive_cases
FROM ctotal AS CT
JOIN cdeath AS CD
  ON CT.country = CD.country AND CT.continent = CD.continent
ORDER BY CT.continent;

SELECT *, ROUND(CAST(active_cases AS float)/CAST(population AS float)*100,2) AS percentofrollingpop
FROM tempactivepop;
```

CREATE MULTIPLE VIEWS

```python
%%sql

DROP VIEW IF EXISTS tempactivepop1;

CREATE VIEW tempactivepop1 AS
SELECT CT.country, CT.continent, CT.population, CD.active_cases, SUM(CAST(CD.active_cases AS int)) OVER (PARTITION BY CT.continent ORDER BY CT.continent, CT.country) AS rollingactive_cases
FROM ctotal AS CT
JOIN cdeath AS CD
  ON CT.country = CD.country AND CT.continent = CD.continent
ORDER BY CT.continent;
```

```python
%%sql
SELECT * FROM tempactivepop1;
```

```python
df = pd.read_sql('SELECT * FROM tempactivepop1;', 'sqlite:///temp.db')
```

```python
df.to_csv('tempactivepop1.csv', index=False)
```

```python
%%sql

DROP VIEW IF EXISTS popactivecases ;

CREATE VIEW popactivecases AS
SELECT CT.country, CT.continent, CT.population, CD.active_cases
FROM ctotal AS CT
JOIN cdeath AS CD
    ON CT.country = CD.country 
    AND CT.continent = CD.continent
ORDER BY  CD.active_cases DESC;
```

```python
%%sql
SELECT * FROM popactivecases;
```

```python
df = pd.read_sql('SELECT * FROM popactivecases;', 'sqlite:///temp.db')
```

```python
df.to_csv('popactivecases.csv', index=False)
```

```python
%%sql

DROP VIEW IF EXISTS maxcontinentdeath;

CREATE VIEW maxcontinentdeath AS 

SELECT continent, MAX(CAST(total_deaths AS int)) AS contt_deaths
FROM ctotal
WHERE continent IS NOT NULL
GROUP BY continent
ORDER BY contt_deaths DESC
```

```python
%%sql
SELECT * FROM maxcontinentdeath;
```
